### **Центральная предельная теорема**

В этом задании нужно убедиться, что ЦПТ действительно работает.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

Создайте случайную величину из любого выбранного вами распределения, для разных значений n сгенерируйте 1000 выборок размера n и постройте гистограммы средних этих выборок.

Пробывал ни одно лишь распределение, остановился на экспоненциальном, потому что оно более показательно в условиях задачи.

```python
#scale = 0.5
#rv = sts.chi(3, 0, scale)
rv = sts.expon(loc = 2.0,  scale = 3.0)
samples = rv.rvs(size=1000)
```

Найдём среднее значение выборки:

```python
x = np.linspace(0,2,100)
pdf = rv.pdf(x)
plt.hist(samples, density=True, bins=30, label='sample histogram')
plt.plot(x, pdf, label='theoretical pdf', alpha=0.5)
plt.legend()
plt.ylabel('fraction of samples, $f(x)$')
plt.xlabel('$x$')
plt.show()
```

```python
samples.mean()
```

```python
#параметры запуска: размеры выборок и количество повторов генерации
def sample(n1):
    #n1 = 1000
    samples_count = 1000
    #массивы для сохранения средних по выборкам
    a1 = np.array([])

    for number in range(samples_count):
        # на каждом цикле генерируем выборки разных размеров
        sample1 = rv.rvs(size=n1)
        # считаем среднее по каждой выборке и добавляем в массив средних
        a1 = np.append(a1, sample1.mean())
    return a1
```

```python
#строим графики
#подписи для графиков
def graph (a1, n1):
    res = sample(n1)
    label_1 = 'sample of ' + str(n1) + ' histogram'
    plt.hist(res, density=True, label=label_1)

    plt.legend()
    plt.ylabel('samples means')
    plt.xlabel('$x$')
```

```python
graph (a1, 1)
```

Построим гистограммы на отдельных  графиках (n меняется от 1 до 10):

```python
#параметры запуска: размеры выборок и количество повторов генерации
n1 = 10
n1_list = list(range(1,n1+1)) # YOUR CODE HERE

for n1 in n1_list:
    plt.figure(figsize=(5,4))
    graph(a1, n1)
```

Построим гистограммы на одном графике (n меняется от 1 до 10):

```python
#параметры запуска: размеры выборок и количество повторов генерации
n1 = 10
n1_list = list(range(1,n1+1)) # YOUR CODE HERE

for n1 in n1_list:
    #plt.figure(figsize=(5,4))
    graph(a1, n1)
```

Построим гистограммы на одном графике для n = 1, 10, 100:

```python
graph(a1,1)
graph(a1,10)
graph(a1,100)
```

Видно, что при увеличении числа выборок n зависимость распределения сужается согласно закону центральной предельной теоремы вокруг среднего значения выборки.  

При росте n распределение случайной величины становится похожей на нормальное распределение с параметрами:  
- среднее значение равно среднему значению случайной величины  
- дисперсия равна дисперсии случайно величины делённое на n.
